````
# импортируем необходимые библиотеки
# для визуализации:
import plotly
import plotly.graph_objects as go 
import plotly.express as px
from plotly.subplots import make_subplots
# для работы с данными:
import numpy as np # для матриц, матриц и математики
import pandas as pd # для таблиц
````

````
# сгенерируем данные для визуализации
# создаём линейное пространство из 200 чисел, 
# которые равномерно распределены на промежутке от 1 до 20 включительно
x = np.linspace(1,20,200)

# напишите функцию, которая будет возводить вектор линейного пространства 
# в квадрат с помощью numpy.square
def squares(x):
  # Ваш код здесь
  return np.square(x)
````

````
# постройте scatter-plot 

fig = go.Figure()
# Ваш код здесь
fig.add_trace(go.Scatter(x=x,y=squares(x)))
fig.show()
````

````
def sin(x, h=False):
  if h: return np.sinh(x)
  else: return np.sin(x)

# второй аргумент "h" - опциональный
# вы можете вызвать функции не указывая его явно
# тогда вызовется условие else

def cos(x, h=False):
  if h: return np.cosh(x)
  else: return np.cos(x)

def tan(x, h=False):
  if h: return np.tanh(x)
  else: return np.tan(x)

def ctan(x, arc=False):
  if arc: return np.arccos(x)/np.arcsin(x)
  else: return np.cos(x)/np.sin(x)


fig = make_subplots(rows=3,cols=2,
                    specs = [[{},{}],[{'colspan':2},None],[{},{}]],
                    subplot_titles=('Plot 1','Plot 2','Plot 3','Plot 4','Plot 5')
                    )

fig.add_trace(go.Scatter(x=x,y=sin(x),name='f(x)=sin(x)'),1,1)
fig.add_trace(go.Scatter(x=x,y=cos(x),name='f(x)=cos(x)'),1,1)

fig.add_trace(go.Scatter(x=x,y=sin(x,h=True),name='f(x)=sinh(x)'),1,2)
fig.add_trace(go.Scatter(x=x,y=cos(x,h=True),name='f(x)=cosh(x)'),1,2)

fig.add_trace(go.Scatter(x=x,y=tan(x),name='f(x)=tan(x)'),2,1)
fig.add_trace(go.Scatter(x=x,y=ctan(x),name='f(x)=ctan(x)'),2,1)

fig.add_trace(go.Scatter(x=x,y=tan(x,h=True),name='f(x)=tanh(x)'),3,1)

fig.add_trace(go.Scatter(x=x,y=squares(x),name='f(x)=x<sup>2</sup>'),3,2)
fig.add_trace(go.Scatter(x=x,y=x,name='f(x)=x'),3,2)


fig.update_layout(legend_orientation='h',
                  margin=dict(l=0,r=20,t=40,b=0),
                  legend=dict(x=.48,xanchor='center'),
                  title='Assignment 2',
                  height=800,
                  hovermode='x')

fig.update_xaxes(title='X1',col=1,row=1)
fig.update_yaxes(title='Y1',col=1,row=1)
fig.update_xaxes(title='X2',col=2,row=1)
fig.update_yaxes(title='Y2',col=2,row=1)

fig.update_xaxes(title='X3',col=1,row=2)
fig.update_yaxes(title='Y3',col=1,row=2)

fig.update_xaxes(title='X4',col=1,row=3)
fig.update_yaxes(title='Y4',col=1,row=3)

fig.update_xaxes(title='X5',col=2,row=3)
fig.update_yaxes(title='Y5',col=2,row=3)


fig.update_traces(hoverinfo='all',
                  hovertemplate='Arg: %{x}<br>Fun: %{y}')
fig.show()
````

````
# сгенерируем данные, имитирующие бросок двух костей 

data = np.random.randint(1,7,size=(100,2))
dices = pd.DataFrame(data=data,columns=('Dice 1','Dice 2'))
dices['sum'] = dices['Dice 1'] + dices['Dice 2']

dices # посмотрите на полученную таблицу
dices.describe() # здесь можно посмотреть сводку о численных данных

sum_counts = dices['sum'].value_counts().sort_index()
sum_counts

labels = ["Всего событий: " + str(sum(sum_counts))]
parents = [""]
values = [sum(sum_counts)]

# 1-й уровень, центр диаграммы
labels = ["Всего событий: " + str(sum(sum_counts))]
parents = [""]
values = [sum(sum_counts)]

# 2-й уровень, "промежуточный"
second_level_dict = {x:'Событий: ' + str(sum_counts[x]) + '<br>Σ = ' + str(x) for x in sum_counts.index}
labels += map(lambda x: second_level_dict[x], sum_counts.index)
parents += [labels[0]]*len(sum_counts)
values += sum_counts.tolist()

# Готовим DataFrame для 3 уровня (группируем )
third_level = dices.groupby(['Dice 1', 'Dice 2']).count().reset_index()
third_level.rename(columns={'sum':'Value'}, inplace=True)
third_level['sum'] = third_level['Dice 1'] + third_level['Dice 2']
third_level['Label'] = third_level['Dice 1'].map(str) + ' + ' + third_level['Dice 2'].map(str)
third_level['Parent'] = third_level['sum'].map(lambda x: second_level_dict[x])
# 3-й уровень, "лепестки" диаграммы
values += third_level['Value'].tolist()
parents += third_level['Parent'].tolist()
labels += third_level['Label'].tolist()

fig = go.Figure(go.Sunburst(
    labels = labels,
    parents = parents,
    values=values,
    branchvalues="total"
))
fig.update_layout(margin = dict(t=0, l=0, r=0, b=0))

fig.show()
````
